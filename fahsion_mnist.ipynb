```python
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow import keras
```

```python
data = keras.datasets.fashion_mnist
(train_X, train_y), (test_X, test_y) = data.load_data()
```

```python
train_X, test_X = train_X/255, test_X/255
```

```python
label = {'0': 	'T-shirt/top', '1': 	'Trouser', '2': 	'Pullover', '3': 	'Dress', '4': 	'Coat', '5': 	'Sandal', '6': 	'Shirt', '7': 	'Sneaker', '8': 	'Bag', '9': 	'Ankle boot'}
```

```python
def show_pic(X, y, start = 1):
  for i in range(start, start + 9):
    plt.subplot(3,3,i)
    plt.gcf().set_size_inches(5, 5)
    plt.gca().axis('off')
    plt.imshow(X[i])
    plt.title(label[str(y[start])])
```

```python
show_pic(train_X, train_y)
```

```python
model = keras.models.Sequential([keras.layers.Flatten(),
                                keras.layers.Dense(28, activation = tf.nn.relu),
                                keras.layers.Dense(10, activation = tf.nn.softmax)])
```

```python
model.compile(optimizer = tf.optimizers.Adam(), loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(train_X, train_y, epochs = 10)
```

```python
model.evaluate(test_X, test_y)
```

```python
predicted = model.predict(test_X[:20])
```

```python
print(list(tf.argmax(predicted, axis = 1).numpy()))
print(list(test_y[:20]))
```
